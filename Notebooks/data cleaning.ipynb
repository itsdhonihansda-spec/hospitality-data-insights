# Check Missing Values

# Check PII Columns

```python
pii_columns = [
    "name",
    "email",
    "phone-number",
    "credit_card"
]

df[pii_columns].head()
```

## Create a Copy

```python
df_clean = df.copy()
```

## Remove PII Columns

```python
pii_columns = [
    "name",
    "email",
    "phone-number",
    "credit_card"
]

df_clean.drop(columns=pii_columns, inplace=True)
```

# Remove company

```python
df_clean.drop(columns=["company"], inplace=True)
```

# Fill Missing Values

```python
df_clean["agent"] = df_clean["agent"].fillna(0)

df_clean["country"] = df_clean["country"].fillna("Unknown")

df_clean["children"] = df_clean["children"].fillna(0)
```

## Convert Data Types

```python
df_clean["children"] = df_clean["children"].astype(int)

df_clean["agent"] = df_clean["agent"].astype(int)
```

# Verify Missing Values

```python
df_clean.isnull().sum().sort_values(ascending=False)
```

## Save the Cleaned Dataset

```python
df_clean.to_csv(
    "../Data/cleaned/hotel_bookings_cleaned.csv",
    index=False
)
```
