# Hospitality Business Intelligence & Revenue Analytics

## Exploratory Data Analysis (EDA)

### Objective

Analyze hotel booking patterns, cancellations, occupancy, revenue, and customer behavior to provide business recommendations.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("ggplot")

pd.set_option("display.max_columns", None)
```

# Load Cleaned Dataset

```python
df = pd.read_csv("../Data/cleaned/hotel_bookings_cleaned.csv")

df.head()
```

# Feature Engineering

```python
df["total_guests"] = (
    df["adults"] +
    df["children"] +
    df["babies"]
)
```

# Total Nights

```python
df["total_nights"] = (
    df["stays_in_weekend_nights"] +
    df["stays_in_week_nights"]
)
```

# Total Revenue

```python
df["total_revenue"] = (
    df["adr"] *
    df["total_nights"]
)
```

# Executive KPI Summary

```python
print("Total Bookings :", len(df))
print("Cancelled Bookings :", df["is_canceled"].sum())
print("Cancellation Rate :", round(df["is_canceled"].mean()*100,2),"%")
print("Average ADR :", round(df["adr"].mean(),2))
print("Estimated Revenue :", round(df["total_revenue"].sum(),2))
```

 # Business Question 1

 Which hotel generates the highest revenue ?

```python
hotel_revenue = (
    df.groupby("hotel")["total_revenue"]
      .sum()
      .sort_values(ascending=False)
)

hotel_revenue
```

# Business Question 2

Which market segment generates the highest revenue ?

```python
segment = (
    df.groupby("market_segment")
      .agg(
          Bookings=("hotel","count"),
          Revenue=("total_revenue","sum")
      )
      .sort_values(
          by="Revenue",
          ascending=False
      )
)

segment
```

# Business Question 3

Which countries generate the most bookings ?

```python
country = (
    df["country"]
      .value_counts()
      .head(10)
)

country
```

# Business Question 4

Which month receives the most bookings ?

```python
month = (
    df["arrival_date_month"]
      .value_counts()
)

month
```

# Business Question 5

Cancellation Rate by Hotel

```python
cancel = (
    df.groupby("hotel")["is_canceled"]
      .mean()*100
)

cancel
```
